```python
import requests
from bs4 import BeautifulSoup

def thesaurus_dot_com(word):
    url = 'https://www.thesaurus.com/browse/{}'.format(word)
    r = requests.get(url)
    returned_words_list = []

    soup = BeautifulSoup(r.text, 'html.parser')
    word_ul = soup.find("ul", {"class":'css-1lc0dpe et6tpn80'})
    for idx, elem in enumerate(word_ul.findAll("a")):
        returned_words_list.append(elem.text.strip())
    return returned_words_list
```

```python
def wordreference(word):
    url = 'https://www.wordreference.com/synonyms/{}'.format(word)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    returned_words_list = []
    word_ul = soup.find('div', style = {"margin-left: 40px;"})
    for idx, elem in enumerate(word_ul.findAll("span")):
        returned_words_list.append(elem.text.strip())
    return returned_words_list

```

```python
def merriam_webster(word):
    url = 'https://www.merriam-webster.com/thesaurus/{}'.format(word)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    #print(soup.prettify())
    returned_words_list = []
    word_ul = soup.find('div', {'class': "thes-list-content synonyms_list"})
    for idx, elem in enumerate(word_ul.findAll("a")):
        returned_words_list.append(elem.text.strip())
    returned_words_list
```
